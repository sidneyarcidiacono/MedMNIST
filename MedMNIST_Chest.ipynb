```
ChestMNIST = np.load("chestmnist.npz")
print(ChestMNIST)
```

```
print(list(ChestMNIST.keys()))
```

```
class_names = ["atelectasis", "cardiomegaly", "effusion", "infiltration", "mass","nodule","pneumonia", "pneumothorax","consolidation","edema",
               "emphysema","fibrosis","pleural","hernia"]
```

```
train_images = ChestMNIST['train_images']
train_labels = ChestMNIST['train_labels'].squeeze()

val_images = ChestMNIST['val_images']
val_labels = ChestMNIST['val_labels'].squeeze()

test_images = ChestMNIST['test_images']
test_labels = ChestMNIST['test_labels'].squeeze()
```

```
train_images = train_images / 255.0

test_images = test_images / 255.0
```

```
plt.figure(figsize=(10,10))
for i in range(25):
  for j in range(len(class_names)):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i][j]])
plt.show()
```
